```python


import basedosdados as bd

billing_id = 'primeiroprojeto-469222'

query = """
  SELECT
    dados.ano as ano,
    dados.id_municipio AS id_municipio,
    diretorio_id_municipio.nome AS id_municipio_nome,
    dados.localizacao as localizacao,
    dados.taxa_abandono_ef_anos_iniciais as taxa_abandono_ef_anos_iniciais
FROM `basedosdados.br_inep_indicadores_educacionais.municipio` AS dados
LEFT JOIN (SELECT DISTINCT id_municipio,nome  FROM `basedosdados.br_bd_diretorios_brasil.municipio`) AS diretorio_id_municipio
    ON dados.id_municipio = diretorio_id_municipio.id_municipio
"""

df = bd.read_sql(query = query, billing_project_id = billing_id)

df
```

```python

# Filtrar df apenas para 2021
import pandas as pd

# 1. Filtrar apenas os dados de 2021
df_2021 = df[df["ano"] == 2021]

# 2. Selecionar apenas as colunas necessárias e remover duplicatas
df_abandono = df_2021[["id_mundv", "taxa_abandono_ef_anos_iniciais"]].drop_duplicates()

# 3. Garantir que as chaves tenham o mesmo tipo
file["id_mundv"] = file["id_mundv"].astype(str)
df_abandono["id_mundv"] = df_abandono["id_mundv"].astype(str)

# 4. Fazer o merge mantendo apenas as linhas do 'file'
df_final = pd.merge(file, df_abandono, on="id_mundv", how="left")

# 5. Conferir o resultado
df_final  # Deve ser (27, ...), ou seja, 27 linhas


```

```python
# Agrupar para garantir apenas uma linha por id_mundv
df_abandono = df_abandono.groupby("id_mundv", as_index=False)["taxa_abandono_ef_anos_iniciais"].first()

# Merge mantendo só as linhas do file
df_final = pd.merge(file, df_abandono, on="id_mundv", how="left")

df_final  
```

```python
# 1. Selecionar colunas importantes
df_abandono = df[["id_mundv", "ano", "taxa_abandono_ef_anos_iniciais"]]

# 2. Garantir mesmo tipo de chave
file["id_mundv"] = file["id_mundv"].astype(str)
df_abandono["id_mundv"] = df_abandono["id_mundv"].astype(str)

# 3. Agrupar por id_mundv e ano para remover duplicatas
df_abandono = df_abandono.groupby(["id_mundv", "ano"], as_index=False).mean()

# 4. Pivot para ter cada ano como coluna
df_abandono_wide = df_abandono.pivot(index="id_mundv", 
                                     columns="ano", 
                                     values="taxa_abandono_ef_anos_iniciais").reset_index()

# 5. Renomear colunas para ficar mais claro
df_abandono_wide.columns = ["id_mundv"] + [f"taxa_abandono_{ano}" for ano in df_abandono_wide.columns[1:]]

# 6. Merge com file mantendo só as 27 linhas
df_final = pd.merge(file, df_abandono_wide, on="id_mundv", how="left")

df_final
```
