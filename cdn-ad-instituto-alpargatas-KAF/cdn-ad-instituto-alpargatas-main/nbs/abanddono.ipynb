```python
import pandas as pd

# Carregar os dados
df = pd.read_excel(
    r'C:\Users\anton\Desktop\data_analise_dados\resultado_final_PB_com_abandono.xlsx',
    sheet_name="Sheet1"
)

# Selecionar colunas relevantes
df_base = df[[
    "ds_mun_ia", "ideb_2019", "ideb_2023",
    "taxa_abandono_2019", "taxa_abandono_2022",
    "nbeneficiados_2025", "nprojetos_2025"
]].drop_duplicates()

# Criar variável de evolução (delta)
df_base["delta_ideb"] = df_base["ideb_2023"] - df_base["ideb_2019"]
df_base["delta_abandono"] = df_base["taxa_abandono_2022"] - df_base["taxa_abandono_2019"]

# Critério de intensidade: acima da mediana de beneficiados em 2025
mediana_beneficiados = df_base["nbeneficiados_2025"].median()
df_base["grupo_intensidade"] = df_base["nbeneficiados_2025"].apply(
    lambda x: "Alta atuação" if x >= mediana_beneficiados else "Baixa atuação"
)

# Calcular médias por grupo
resultado_grupos = df_base.groupby("grupo_intensidade")[["delta_ideb", "delta_abandono"]].mean().reset_index()

print("📊 Evolução média por grupo de intensidade (antes e depois):")
print(resultado_grupos)

# Exportar resultado
resultado_grupos.to_excel(
    r'C:\Users\anton\Desktop\data_analise_dados\resultado_grupos_intensidade.xlsx',
    index=False
)
```

```python
import pandas as pd

# Carregar os dados
df = pd.read_excel(
    r'C:\Users\anton\Desktop\data_analise_dados\resultado_final_PB_com_abandono.xlsx',
    sheet_name="Sheet1"
)

# Selecionar colunas relevantes e remover duplicadas por município
df_base = df[[
    "ds_mun_ia", "ideb_2019", "ideb_2023",
    "taxa_abandono_2019", "taxa_abandono_2022",
    "nbeneficiados_2025", "nprojetos_2025"
]]

# Criar variável de evolução (delta)
df_base["delta_ideb"] = df_base["ideb_2023"] - df_base["ideb_2019"]
df_base["delta_abandono"] = df_base["taxa_abandono_2022"] - df_base["taxa_abandono_2019"]

# Critério de intensidade: acima da mediana de beneficiados em 2025
mediana_beneficiados = df_base["nbeneficiados_2025"].median()
df_base["grupo_intensidade"] = df_base["nbeneficiados_2025"].apply(
    lambda x: "Alta atuação" if x >= mediana_beneficiados else "Baixa atuação"
)

# Calcular médias por grupo
resultado_grupos = df_base.groupby("grupo_intensidade")[["delta_ideb", "delta_abandono"]].mean().reset_index()

# Juntar as médias ao lado de cada município
df_final = df_base.merge(
    resultado_grupos,
    on="grupo_intensidade",
    suffixes=("", "_media_grupo")
)

# Selecionar colunas finais
df_final = df_final[[
    "ds_mun_ia", "grupo_intensidade", "delta_ideb", "delta_ideb_media_grupo",
    "delta_abandono", "delta_abandono_media_grupo"
]]

print(df_final)
```
