```python




# Filtrar apenas as linhas para Alagoa Nova e rede "Pública"
df_alagoa_publica = df[(df['ds_mun_ia'] == 'ALAGOA NOVA') & (df['rede'] == 'Pública')].copy()

# Selecionar colunas de notas de Matemática e Português
nota_cols = {
    'matematica': ['VL_NOTA_MATEMATICA_2005', 'VL_NOTA_MATEMATICA_2007', 'VL_NOTA_MATEMATICA_2009',
                   'VL_NOTA_MATEMATICA_2011', 'VL_NOTA_MATEMATICA_2013', 'VL_NOTA_MATEMATICA_2015',
                   'VL_NOTA_MATEMATICA_2017', 'VL_NOTA_MATEMATICA_2019', 'VL_NOTA_MATEMATICA_2021',
                   'VL_NOTA_MATEMATICA_2023'],
    'portugues': ['VL_NOTA_PORTUGUES_2005', 'VL_NOTA_PORTUGUES_2007', 'VL_NOTA_PORTUGUES_2009',
                  'VL_NOTA_PORTUGUES_2011', 'VL_NOTA_PORTUGUES_2013', 'VL_NOTA_PORTUGUES_2015',
                  'VL_NOTA_PORTUGUES_2017', 'VL_NOTA_PORTUGUES_2019', 'VL_NOTA_PORTUGUES_2021',
                  'VL_NOTA_PORTUGUES_2023']
}

# Extrair os anos
anos = [int(col.split('_')[-1]) for col in nota_cols['matematica']]

# Converter colunas de notas para numérico, lidando com valores ausentes
for disciplina in ['matematica', 'portugues']:
    df_alagoa_publica[nota_cols[disciplina]] = df_alagoa_publica[nota_cols[disciplina]].apply(pd.to_numeric, errors='coerce')

# Pegar os dados das notas
notas_matematica = df_alagoa_publica[nota_cols['matematica']].iloc[0]
notas_portugues = df_alagoa_publica[nota_cols['portugues']].iloc[0]

# Criar o gráfico de barras agrupadas
fig, ax = plt.subplots(figsize=(12, 6))

bar_width = 0.35
index = range(len(anos))

plt.bar(index, notas_matematica, bar_width, label='Matemática', color='#FF9999')
plt.bar([i + bar_width for i in index], notas_portugues, bar_width, label='Português', color='#66B2FF')

plt.xlabel('Ano')
plt.ylabel('Nota')
plt.title('Evolução das Notas de Matemática e Português - Alagoa Nova (Rede Pública)')
plt.xticks([i + bar_width/2 for i in index], anos, rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
